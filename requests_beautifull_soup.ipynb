# working with requests and beautifull soup

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://pythonhow.com/example.html")
c = r.content
```

```python
c
```

### use bs and html parser to read content stored in c

```python
soup = BeautifulSoup(c, "html.parser")
```

```python
print(soup.prettify())
```

### example,select div class cities

```python
all = soup.find_all("div", {"class":"cities"})
all
```

### working example, select p element

```python
all2 = soup.find("p")
all2
```
